```python
import numpy as np
```

```python
x = np.linspace(0,5,11)

y = x**2
```

```python
    #FUNCTINAL MENTHOD TO CREATE PLOT 
```

```python
plt.plot(x,y)
plt.xlabel('X Label')#to add the label in the graph
plt.ylabel('Y Label')#to add the label in the graph
plt.title("title ")#to write the title of graph
plt.show()
```

```python
plt.subplot(1,2,2)  # 1.st Row , No of column , plot no refering to 
plt.plot(x,y,'r') # 'r' MEANS COLOR OF PLOT


plt.subplot(1,2,1)
plt.plot(y,x,'b')# 'r' MEANS COLOR OF PLOT bLUE
```

```python
#OBJECT ORIENTED METHOD 
```

```python
fig = plt.figure()#figure oobject 

axes= fig.add_axes([0.1,0.1,0.8,0.8])#LIST FOUR LEFT BOTTOM AND WIDTH AND HEIGHT ,method on figure object

axes.plot(x,y)

axes.set_xlabel('X label')

axes.set_xlabel('Y label')

axes.set_xlabel('Set tITILE ')

```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)

axes1.set_title('Larger plot')
axes2.plot(y,x)

axes2.set_title('Small plot')



```
