# Bar chart 

```python
df1 = pd.read_csv('C:\\Users\\dt321\\Desktop\\PythonLibrary\\DataVisualization\\matplotlib\\music.csv',index_col='Year')
```

```python
df1.head()
```

```python
# initialize the graph
```

```python
fig1, ax1 = plt.subplots(figsize=(12,4))


rock=ax1.bar(df1.index,'Rock',data= df1 ,label ='% Rock')

jazz=ax1.bar(df1.index,'Jazz',data= df1 ,label = '% Jazz')

ax1.legend()

ax1.set_title('Rock vs Jazz',size =23)

ax1.set_xticks(df1.index)

spines=['bottom','top','left','right']


for s in spines:
    
    if s in ('top','right'):
        
        ax1.spines[s].set_visible(False)
        
ax1.legend(frameon= False)

ax1.autoscale()





```

```python
for p in rock.patches:
    
    t= ax1.annotate(str(p.get_height())+'%',(p.get_x(),p.get_height()+2))
    t.set(color='black',size=8)
    
    t.set_bbox(dict(facecolor='White',boxstyle='circle',lw=1,edgecolor='black'))
    
fig1
```

# Scatter plot

```python
df3 = pd.read_csv("C:\\Users\\dt321\\Desktop\\PythonLibrary\\DataVisualization\\matplotlib\\scatter.csv",index_col='Year')
```

```python
df3.head()
```

```python
a='Federal Funds Target Rate'
b= 'Unemployment Rate'

fig4,ax4 =plt.subplots(figsize = (12,4))

p=ax4.scatter(a,b,data=df3)
```

```python
help(p)
```

```python
ax4.set_title("this is scatter plot ",size =22)

ax4.set_xlabel("interest rate ",size =13)
ax4.set_ylabel("Unemployement rate ",size =13)

from matplotlib.ticker import StrMethodFormatter

mft1=StrMethodFormatter("{x}%")

ax4.xaxis.set_major_formatter(mft1)

ax4.yaxis.set_major_formatter(mft1)

fig4
```

```python
p.set_alpha(0.6)

p.set_sizes(df3[a].dropna()**2.8)

p.set_color(['White','yellow'])

spince = ['top','bottom','left','right']


for s in spince:
    
    ax4.spines[s].set_visible(False)

ax4.set_facecolor('#01599b')

fig4.set_facecolor('#01599b')

ax4.title.set_color('White')

ax4.tick_params(color="white")


ax4.xaxis.get_label().set_color('White')

ax4.xaxis.label.set_color('White')


fig4
```
