# Add line 2D artist

```python
import numpy as np

from matplotlib.lines import Line2D

import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
np.random.seed(15)
```

```python
x = np.linspace(1,40,10)#1 start  till 40 gives 15 random number 
y1= x/(np.random.rand(10)*2)
y2= x/(np.random.rand(10)*3)
```

```python
line1 = Line2D(x,y1)
line2 = Line2D(x,y2)
```

```python
fig, ax = plt.subplots()
ax.add_line(line1)
ax.add_line(line2)
```

```python
ax.autoscale()
```

```python
fig
```

# size varioation  of plot 

```python
# help(fig)
```

```python
fig.set_size_inches(25,10)
```

```python
fig
```

```python
l1=ax.lines[0]
l2=ax.lines[1]
```

```python
# help(l1)
```

```python
l1.set_color('r')
fig
```

```python
l1.set_alpha(0.3)
fig
```

```python
for line in ax.lines:
    line.set_color('r')
    line.set_lw(2)
```

```python
fig
```

```python
l1.set_linestyle('dashed')
```

```python
l2.set_marker('*')
l2.set_markersize(20)
l2.set_markerfacecolor("white")
l2.set_markevery(1)
l2.set_sketch_params(0.9)
```

```python
fig
```

```python
l1.set_clip_on(False)
ax.set_xlim(0,39)
fig
```

```python
ax.set_title('Line chart ')
ax.set_xlabel('This is X axis ')
ax.set_ylabel('This is y axis ')
```

```python
fig 
```

```python
fig.savefig('artist.png',dpi= 350,transparent=False)
```

# Add reactangle patches

```python
from matplotlib.patches import Rectangle
```

```python
react1 = Rectangle((0.5,0),0.2,0.8)
react2 = Rectangle((0.2,0),0.2,0.8)
```

# Instanciate figure

```python
fig1,ax3 = plt.subplots()

ax3.add_patch(react1)
ax3.add_patch(react2)
```

```python
ax3.patches
```

# Propertiese of reactange 

```python
x= list(range(1,6))
y = np.linspace(1,10,5)
fig6,ax6 = plt.subplots()

def makebar (ax6,x,y,w):
    
    for x,y in zip(x,y):
        
        bar = Rectangle((x,0),w,y)
        
        ax6.add_patch(bar)

makebar(ax6,x,y,0.4)


ax6.autoscale()

fig6.set_size_inches(15,5)

```

```python
ax6.set_title('Line chart ')
ax6.set_xlabel('This is X axis ')
ax6.set_ylabel('This is y axis ')
```
