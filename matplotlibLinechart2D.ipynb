# Fancy Bbox patch

```python
from matplotlib.patches import FancyBboxPatch
```

```python
help(FancyBboxPatch)
```

```python
x = list (range(1,10))
y = x
fig7 , ax7 = plt.subplots(figsize = (15,5))
def makeFancyBar(ax7,x,y,w):
    
    for x,y in zip(x,y):
        
        fbox = FancyBboxPatch((x,0),w,y,boxstyle='round')
        ax7.add_patch(fbox)
    
makeFancyBar(ax7,x,y,0.3)

ax7.autoscale()

```

```python
box_style = ['circle','darrow','larrow','rarrow','round4','roundtooth','sawtooth','square','circle','round']

for p ,s in zip(ax7.patches,box_style):
    
    p.set_boxstyle(s)

fig7
```

# Add legend and Legend artist

```python
import matplotlib.pyplot as plt 

from matplotlib.legend import Legend

from matplotlib.patches import Rectangle

%matplotlib inline
```

```python
reac1 = Rectangle((0.1,0),0.1,0.3,label='Day 1')
reac2 = Rectangle((0.3,0),0.1,0.6,label='Day 2')
reac3 = Rectangle((0.7,0),0.1,0.9,label='Day 3')

fig11,ax11 = plt.subplots()

ax11.add_patch(reac1)
ax11.add_patch(reac2)
ax11.add_patch(reac3)

color =['red','green','black']

for p , c in zip(ax11.patches,color):
    
    p.set_color(c)


handles ,labels = ax11.get_legend_handles_labels()

legend = Legend(ax11,handles=handles,labels=labels)

ax11.add_artist(legend)
```

```python
ax11.legend()
fig11
```

# Legend Properties 

```python
legend.texts
```

```python
legend.texts[1].set(color='black',alpha=0.5)

fig11.set_size_inches(15,5)
fig11
```

```python
legend.legendHandles[2].set_facecolor("Red")
legend.legendHandles[2].set_alpha(0.2)
```
